```python
def load_h5(h5_filename):
    f = h5py.File(h5_filename)
    data = f['data'][:]
    label = f['label'][:]
    return (data, label)
```

```python
load_h5(r'C:\Users\julcs\Pointcloud\ply_data_test0.h5')
```

```python
def load_h5_all(h5_filename):
    f = h5py.File(h5_filename)
    data = f['data'][:]
    label = f['label'][:]
    ID = f['faceId'][:]
    normal= f['normal'][:]
    return (data, label,ID, normal)
```
